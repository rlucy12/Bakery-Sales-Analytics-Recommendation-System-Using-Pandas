```python
# Market Basket analysis (Paring Matrix)
```

```python
pairing_matrix = pd.DataFrame()

for item in item_columns:
    
    item_orders = data[data[item] > 0]
    
    pairing = item_orders[item_columns].sum()
    
    pairing_matrix[item] = pairing

# Remove self pairing
for item in item_columns:
    pairing_matrix.loc[item, item] = 0

print(pairing_matrix)
```

```python
basket_distribution = data['basket_size'].value_counts().sort_index()
basket_percentage = (basket_distribution / len(data)) * 100

basket_summary = pd.DataFrame({
    'Order_Count': basket_distribution,
    'Percentage (%)': basket_percentage.round(2)
})

print("\nCorrect Basket Summary:")
print(basket_summary)
```

```python
combo_pairs = []

for item in pairing_matrix.columns:
    
    top_pairs = pairing_matrix[item].sort_values(ascending=False).head(3)

    for pair_item, score in top_pairs.items():
        
        combo_pairs.append({
            'Product_A': item,
            'Product_B': pair_item,
            'Co_Occurrence': score
        })

combo_df = pd.DataFrame(combo_pairs)
```

```python
# Cart Recommendation System
```

```python
def recommend_for_cart(cart_items, top_n=3):

    valid_items = [item for item in cart_items if item in pairing_matrix.columns]

    if not valid_items:
        return "No valid products in cart"

    combined_scores = pairing_matrix[valid_items].sum(axis=1)

    # remove already selected products
    combined_scores = combined_scores.drop(valid_items)

    return combined_scores.sort_values(ascending=False).head(top_n)
```

```python
print(recommend_for_cart(['croissant']))
```

```python
user_input = input("Enter products separated by comma: ")

cart_items = [item.strip() for item in user_input.split(',')]

print(recommend_for_cart(cart_items))
```

```python
# Morning Combo Analysis
```

```python
morning_orders = data[(data['hour'] >= 9) & (data['hour'] <= 12)]

morning_pairing = pd.DataFrame()

for item in item_columns:

    item_orders = morning_orders[morning_orders[item] > 0]

    pairing = item_orders[item_columns].sum()

    morning_pairing[item] = pairing

for item in item_columns:
    morning_pairing.loc[item, item] = 0

print("Morning Combo Candidates")
print(morning_pairing.sum().sort_values(ascending=False).head(5))
```

```python
# A products → Promote & highlight on menu

# B products → Bundle with A products

# C products → Discount or remove
```

```python
# Calculate total quantity sold per product
product_sales = data[item_columns].sum()

# Create DataFrame
abc_df = pd.DataFrame({
    'Product': product_sales.index,
    'Total_Sales': product_sales.values
})

# Sort by sales
abc_df = abc_df.sort_values(by='Total_Sales', ascending=False)

# Calculate cumulative percentage
abc_df['Cumulative_%'] = abc_df['Total_Sales'].cumsum() / abc_df['Total_Sales'].sum() * 100

# Assign ABC category
def classify_abc(x):
    if x <= 70:
        return 'A'
    elif x <= 90:
        return 'B'
    else:
        return 'C'

abc_df['Category'] = abc_df['Cumulative_%'].apply(classify_abc)

print("ABC Product Classification")
print(abc_df)
```

```python
# Demand Trend Detection
```

```python
# Rising Star → Increase inventory

# Declining → Reduce stock or promote

# Stable → Keep normal production
```

```python
monthly_sales = data.groupby(['year','month'])[item_columns].sum().reset_index()

trend_results = {}

for item in item_columns:

    monthly_values = monthly_sales[item]

    growth = monthly_values.diff().mean()

    if growth > 0:
        trend_results[item] = "Rising Star"

    elif growth < 0:
        trend_results[item] = "Declining"

    else:
        trend_results[item] = "Stable"

trend_df = pd.DataFrame(
    list(trend_results.items()),
    columns=['Product','Trend']
)

print("Demand Trend Detection")
print(trend_df)
```

```python
# Peak Demand Analysis
```

```python
peak_list = []

for item in item_columns:
    
    if data[item].sum() > 0:
        
        # Peak hour
        hour_sales = data.groupby('hour')[item].sum()
        peak_hour = hour_sales.idxmax()
        
        # Peak day
        day_sales = data.groupby('day_of_week')[item].sum()
        peak_day = day_sales.idxmax()
        
        peak_list.append({
            'Product': item,
            'Peak_Hour': peak_hour,
            'Peak_Day': peak_day
        })

peak_df = pd.DataFrame(peak_list)

print("Peak Demand Alert")
print(peak_df)
```
